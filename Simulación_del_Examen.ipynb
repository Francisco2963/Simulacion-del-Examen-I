```python
import pandas as pd
import numpy as np
import random
```

```python
# Configuramos la semilla para reproducibilidad
np.random.seed(42)
```

```python
# Generamos los datos sintéticos
user_id = range(1, 301)
app_version = np.random.choice(['1.0', '1.1', '1.2', '1.3'], 300)
platform = np.random.choice(['Android', 'iOS'], 300)
session_duration = np.random.randint(1, 180, 300)
number_of_sessions = np.random.randint(1, 20, 300)
country = np.random.choice(['USA', 'Canada', 'Mexico', 'UK', 'Germany', 'France', 'Spain', 'Italy'], 300)
user_feedback = np.random.randint(1, 6, 300)
```

```python
# Creamos el DataFrame
data = pd.DataFrame({
    'user_id': user_id,
    'app_version': app_version,
    'platform': platform,
    'session_duration': session_duration,
    'number_of_sessions': number_of_sessions,
    'country': country,
    'user_feedback': user_feedback
})
```

```python
# Guardamos los datos en un archivo CSV
data.to_csv('user_app_data.csv', index=False)
print("Datos generados y guardados en 'user_app_data.csv'")
```

```python
import matplotlib.pyplot as plt
```

```python
df.head()
```

```python
gb_A1 = df.groupby('app_version')['user_feedback'].value_counts().round()
gb_A1
```

```python
x = gb_A1.index
print(x)
```

```python
y = gb_A1.values
print(y)
```

```python
x = [' '.join(map(str,item)) for item in gb_A1.index]
y = gb_A1.values
plt.figure(figsize=(10,5))
plt.title('Valoracion de la version de la aplicacion')
plt.bar(x,y)
plt.xlabel('app_version')
plt.ylabel('user_feedback')
```

El grafico muestra quue la version de la apñicacion mejor valorada es la 1.3 con 19 usuarios que la calificaron con la nota mas alta (5).

```python
df.head()
```

```python
gb_A2 = df.groupby('country')['user_feedback'].value_counts()
gb_A2
```

```python
x_values = [item[0] for item in gb_A2.index]
y_values = gb_A2.values

plt.figure(figsize=(10,5))
plt.title('Valoracion por pais')
plt.plot(x_values, y_values, 'o-')
plt.grid(axis = 'x')
plt.xlabel('country')
plt.ylabel('user_feedback')
plt.xticks(rotation = 45)
plt.show()
```

El grafico muestra que el Reino Unido ( UK) es el pais donde mas valoraciones positivas entrega a las versiones de la aplicacion.
